```python
-- PENGERTIAN POLYMORPHISM --
Polymorphism adalah kemampuan sebuah object untuk mempunyai banyak bentuk atau 
mengungkapkan banyak hal melalui satu cara yang sama. 
Polymorphism ada dua tipe, yaitu Overloading dan Overriding.
1. Overloading : method memiliki nama yang sama namun parameter method-nya berbeda.
2. Overriding : method mempunyai nama yang sama dan fungsionalitas yang sama. Namun cara kerja fungsionalitas berbeda.

#Example1:
class Bear():
    def sound (self):
        print("Roooaaaarrr")
class Dog():
    def sound(self):
        print("Whoooff")

def makesound(animal):
    animal.sound()
    
BearObj = Bear()
DogObj = Dog()
makesound(BearObj)

#Example2:
class shark():
    def swim(self):
        print("The shark is swimming")
    def swim_backwards(self):
        print("The shark cannot swim backwards, but can sink backwards")
    def skeleton(self):
        print("The shark's skeleton is made of cartilage")

class clownfish():
    def swim(self):
        print("The clownfish is swimming")
    def swim_backwards(self):
        print("The clownfish can swim backwards")
    def skeleton(self):
        print("The clownfish skeleton is made of bone")

sammy = shark()
sammy.swim()
sammy.swim_backwards()
sammy.skeleton()

casey = clownfish()
casey.swim()
casey.swim_backwards()
casey.skeleton()
```

```python
#Polymorphism w/ function.

class shark():
    def swim(self):
        print("The shark is swimming")
    def swim_backwards(self):
        print("The shark cannot swim backwards, but can sink backwards")
    def skeleton(self):
        print("The shark's skeleton is made of cartilage")

class clownfish():
    def swim(self):
        print("The clownfish is swimming")
    def swim_backwards(self):
        print("The clownfish can swim backwards")
    def skeleton(self):
        print("The clownfish skeleton is made of bone")
        
def in_the_pacific(fish):
    fish.swim()
    fish.swim_backwards()
    fish.skeleton()
    
    
sammy = shark()
casey = clownfish()

in_the_pacific(sammy)
in_the_pacific(casey)
```

```python
#Polymorphism w/ class methods

class shark():
    def swim(self):
        print("The shark is swimming")
    def swim_backwards(self):
        print("The shark cannot swim backwards, but can sink backwards")
    def skeleton(self):
        print("The shark's skeleton is made of cartilage")

class clownfish():
    def swim(self):
        print("The clownfish is swimming")
    def swim_backwards(self):
        print("The clownfish can swim backwards")
    def skeleton(self):
        print("The clownfish skeleton is made of bone")
        
sammy = shark()
casey = clownfish()

for fish in (sammy,casey):
    fish.swim()
    fish.swim_backwards()
    fish.skeleton()

```

```python
--- Abstract Class ---
Abstract Class adalah Class yang mempunyai setidaknya satu abstract method.
â€¢ Abstract method adalah method yang tidak memiliki body (hanya deklarasi method) memiliki body (hanya deklarasi method). 
#Example:

class Document:
    def __init__(self, name):    
        self.name = name
 
    def show(self):             
        raise NotImplementedError("Subclass must implement abstract method")
 
class Pdf(Document):
    def show(self):
        return 'Show pdf contents!'
 
class Word(Document):
    def show(self):
        return 'Show word contents!'
 
documents = [Pdf('Document1'),
             Pdf('Document2'),
             Word('Document3')]
 
for document in documents:
    print (document.name + ': ' + document.show())
```
